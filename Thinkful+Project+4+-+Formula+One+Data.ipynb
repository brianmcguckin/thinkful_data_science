```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
races = pd.read_csv('races.csv')
races
```

```python
#scatter plot of championship rounds per year
plt.scatter(races['year'], races['round'])
plt.show()
```

```python
lap_times = pd.read_csv('lapTimes.csv')
lap_times
```

```python
#access only lap times for raceId 1, 2009 Australian Grand Prix
australian_gp_2009 = lap_times.loc[lambda df: lap_times['raceId'] == 1, :]
australian_gp_2009
```

```python
#Fernando Alonso's race data for the 2009 Australian GP
australian_gp_2009_alonso = australian_gp_2009.loc[lambda df: australian_gp_2009['driverId'] == 4, :]
australian_gp_2009_alonso
```

```python
#plot showing Alonso's lap time for each lap
plt.plot(australian_gp_2009_alonso['lap'], australian_gp_2009_alonso['milliseconds'], color='red')
plt.xlabel('Lap Number')
plt.ylabel('Time in Milliseconds')
plt.title('Fernando Alonso\'s 2009 Aus GP Lap Times')
plt.show()
```

-Here are Fernando Alonso's lap times (in milliseconds) for the 2009 Australian Grand Prix

-From the lap tims, we can infer that Alonso likely had a full fuel/tires pitstop around lap 20

-There was likely an accident and safety car deployment shortly after this as there were a few sequential laps with suppressed times

-Alonso likely pitted around lap 51 or 52, either for a new set of tires or a splash of fuel

-Last laps are slow, indicating the race finished under the safety car, which is in fact the case

-Fun fact: the 2009 Australian GP was only the 2nd race in F1 history to end under a safety car

```python
#plot showing Alonso's position for each lap
plt.scatter(australian_gp_2009_alonso['lap'], australian_gp_2009_alonso['position'], color='red')
plt.xlabel('Lap Number')
plt.ylabel('Race Position')
plt.title('Fernando Alonso\'s 2009 GP Aus Race Position')
plt.show()
```

This scatter plot shows Alonso's (ALO) race position for each of the 58 laps

ALO starts the race in P14 and either had a subpar start, driver behind had a great start, or
lost the position in the opening mid pack battle

Right around lap 10 ALO rapidly gains positions, perhaps indicating other cars are pitting

He briefly reaches P5, then drops rapidly, coinciding with his suspected pit from lap times

Between laps ~27 to 50 he gradually gains positions, likely a combination of a fruitful drive
#and other drivers succumbing to attrition

Drops POS again coinciding with his late race pit, but gains them back as competitors also pit

He started the race in P14, ended in P6 with P5 being the highest held POS for ALO

```python
#let's plot Alonso's lap times vs. race winner Jenson Button's lap times

#get a data frame of Jenson Button's race data, driverID 18
australian_gp_2009_button = australian_gp_2009.loc[lambda df: australian_gp_2009['driverId'] == 18, :]
australian_gp_2009_button
```

```python
plt.plot(australian_gp_2009_button['lap'], australian_gp_2009_button['milliseconds'], color='blue')
plt.plot(australian_gp_2009_alonso['lap'], australian_gp_2009_alonso['milliseconds'], color='red')
plt.xlabel('Lap Number')
plt.ylabel('Time in Milliseconds')
plt.title('Jenson Button and Fernando Alono\'s 2009 Aus GP Lap Times')
plt.show
```

Here we see a big difference in lap time under safety car, likely due to Button leading and being directly behind SC
(Cars further back in the field are allowed to drive faster to catch up to the field)

Button's (BUT) first pit stop appears to have coincided with the safety car deployment

BUT second pit stop occured 3 or 4 laps before ALO

Same scenario for lap times under the second and final safety car deployment

Comparing lap times under green flag conditions, it's clear that Button's drive was faster than Alonso's

```python
#plot showing Button versus Alonso's position for each lap
plt.scatter(australian_gp_2009_alonso['lap'], australian_gp_2009_alonso['position'], color='red')
plt.scatter(australian_gp_2009_button['lap'], australian_gp_2009_button['position'], color='blue')
plt.xlabel('Lap Number')
plt.ylabel('Race Position')
plt.title('Jenson Button and Fernando Alonso\'s 2009 Aus GP Race Position')
plt.show()
```

This scatter plot clearly shows that Button held P1 from race start to completion

Anyone who remembers the 2009 GP season can explain how lopsided racing was in Brawn GP's favor for
much of the year and especially the opening rounds due to the double diffuser controversey

```python
#let's put the 2 comparison plots side by side for easier consumption
plt.figure(figsize=(15,7.5)) #define the figure

plt.subplot(1,2,1) #we want to create a 1 high, 2 across plot, this is for the plot in the 1st position
plt.plot(australian_gp_2009_alonso['lap'], australian_gp_2009_alonso['milliseconds'], color='red', label='Alonso')
plt.plot(australian_gp_2009_button['lap'], australian_gp_2009_button['milliseconds'], color='blue', label='Button')
plt.xlabel('Lap Number')
plt.ylabel('Time in Milliseconds')
plt.title('Jenson Button and Fernando Alono\'s 2009 Aus GP Lap Times')
plt.legend()

plt.subplot(1,2,2) #the last 2 indicates that this plot is going in the 2nd position
plt.scatter(australian_gp_2009_alonso['lap'], australian_gp_2009_alonso['position'], color='red', label='Alonso')
plt.scatter(australian_gp_2009_button['lap'], australian_gp_2009_button['position'], color='blue', label='Button')
plt.xlabel('Lap Number')
plt.ylabel('Race Position')
plt.title('Jenson Button and Fernando Alonso\'s 2009 Aus GP Race Position')
plt.legend()


plt.tight_layout() #make it look nice
plt.show()
```

Here we simply have the above plots side by side, for better comparison and compact viewing.

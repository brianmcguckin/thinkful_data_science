```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
variable1 = np.random.beta(1,2,100)

plt.axvline(variable1.mean(), color = 'b', linestyle = 'solid', linewidth = 2)
plt.axvline(variable1.mean() + variable1.std(), color = 'b', linestyle = 'dashed', linewidth = 2)
plt.axvline(variable1.mean() - variable1.std(), color = 'b', linestyle = 'dashed', linewidth = 2)

plt.hist(variable1)
plt.title('Beta Distribution, a = 1 b = 2')
plt.show()
```

```python
exp = np.random.exponential(10,100)

plt.axvline(exp.mean(), color = 'b', linestyle = 'solid', linewidth = 2)
plt.axvline(exp.mean() + exp.std(), color = 'b', linestyle = 'dashed', linewidth = 2)
plt.axvline(exp.mean() - exp.std(), color = 'b', linestyle = 'dashed', linewidth = 2)

plt.hist(exp)
plt.title('Exponential Distribution, scale = 10')
plt.show()
```

```python
pareto = np.random.pareto(10,100)

rand1 = np.random.normal(0,.5,100)

plt.axvline(pareto.mean(), color = 'b', linestyle = 'solid', linewidth = 2)
plt.axvline(pareto.mean() + pareto.std(), color = 'b', linestyle = 'dashed', linewidth = 2)
plt.axvline(pareto.mean() - pareto.std(), color = 'b', linestyle = 'dashed', linewidth = 2)

plt.hist(rand1, alpha = 0.7)
plt.hist(pareto, alpha = 0.7)
plt.title('Pareto distribution, a = 10')
plt.show()
```

```python
geometric = np.random.geometric(.1,100)

plt.axvline(geometric.mean(), color = 'b', linestyle = 'solid', linewidth = 2)
plt.axvline(geometric.mean() + geometric.std(), color = 'b', linestyle = 'dashed', linewidth = 2)
plt.axvline(geometric.mean() - geometric.std(), color = 'b', linestyle = 'dashed', linewidth = 2)

plt.hist(geometric)
plt.title('Geometric distribution, p = 0.1')
plt.show()
```

```python
uniform = np.random.uniform(0,2,100)

plt.axvline(uniform.mean(), color = 'b', linestyle = 'solid', linewidth = 2)
plt.axvline(uniform.mean() + uniform.std(), color = 'b', linestyle = 'dashed', linewidth = 2)
plt.axvline(uniform.mean() - uniform.std(), color = 'b', linestyle = 'dashed', linewidth = 2)

plt.hist(uniform)
plt.title('Uniform distribution, low = 0 high = 20')
plt.show()
```

```python
vonmises = np.random.vonmises(25,2,10000)

plt.axvline(vonmises.mean(), color = 'b', linestyle = 'solid', linewidth = 2)
plt.axvline(vonmises.mean() + vonmises.std(), color = 'b', linestyle = 'dashed', linewidth = 2)
plt.axvline(vonmises.mean() - vonmises.std(), color = 'b', linestyle = 'dashed', linewidth = 2)

plt.hist(vonmises)
plt.title('Von Mises Distribution, mu = 25 kappa = 2')
plt.show()
```

```python
rand1 = np.random.normal(5, .5, 100)
rand2 = np.random.normal(10, 1, 100)
rand3 = rand1 + rand2 #rand1 and rand2 added together

plt.axvline(rand3.mean(), color = 'b', linestyle = 'solid', linewidth = 2)
plt.axvline(rand3.mean() + rand3.std(), color = 'b', linestyle = 'dashed', linewidth = 2)
plt.axvline(rand3.mean() - rand3.std(), color = 'b', linestyle = 'dashed', linewidth = 2)

plt.hist(rand3)
plt.title('Histogram of rand3')
plt.show()
```
